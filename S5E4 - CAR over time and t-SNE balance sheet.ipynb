```python
def convert_dtypes(df):
    def contains_dates(series):
        date_count = 0
        for value in series:
            try:
                pd.to_datetime(value, errors='raise')
                date_count += 1
            except:
                pass
            if date_count >= 2:
                return True
        return False

    def contains_only_letters(series):
        for value in series:
            if not isinstance(value, str) or not value.isalpha():
                return False
        return True

    for column in df.columns:
        # If the column is an object dtype and contains only letters, skip the conversion
        if df[column].dtype == 'object' and contains_only_letters(df[column]):
            continue

        # If the column is an object dtype, try converting it to a numeric dtype
        if df[column].dtype == 'object':
            numeric_column = pd.to_numeric(df[column], errors='ignore')
            
            # Only convert to datetime if numeric conversion is not possible and the column contains dates
            if numeric_column.dtype == 'object' and contains_dates(df[column]):
                df[column] = pd.to_datetime(df[column], errors='ignore')
            else:
                df[column] = numeric_column

    return df
```

## A closer look at Silicon Valley Bank balance sheet information
The source of data is their own balance sheets, which were downloaded from the Silicon Valley Bank website. Plots are normalised to show the percentage contribution of various elements to the total assets and total liabilities

```python
from pathlib import Path
import pandas as pd
import plotly.graph_objects as go

balance_sheet_summary_path = Path("data", "sivb_balance_sheet_summary.csv")
df_sivb_blsh = pd.read_csv(balance_sheet_summary_path, skiprows=2, index_col=0)
df_sivb_blsh = df_sivb_blsh.transpose()
df_sivb_blsh = df_sivb_blsh.sort_index()
# get a nice year column
df_sivb_blsh["Year Ending"] = df_sivb_blsh.index

assets = [
    "Cash and cash equivalents",
    "Available-for-sale securities, at fair value",
    "Held-to-maturity securities",
    "Non-marketable and other equity securities",
    "Net loans",
    "Premises and equipment, net of accumulated depreciation and amortization",
    "Goodwill",
    "Other intangible assets, net",
    "Lease right-of-use assets",
    "Accrued interest receivable and other assets",
]

liabilities = [
    "Noninterest-bearing demand deposits",
    "Interest-bearing deposits",
    "Total deposits",
    "Short-term borrowings",
    "Lease liabilities",
    "Other liabilities",
    "Long-term debt",
]
```

```python
# normalise assets by total assets 
for asset in assets:
    df_sivb_blsh[asset] = (df_sivb_blsh[asset] / df_sivb_blsh["Total assets"])*100

# draw an area graph
fig = go.Figure()
for asset in assets:
    fig.add_trace(go.Scatter(
        x=df_sivb_blsh["Year Ending"], y=df_sivb_blsh[asset],
        hoverinfo='x+y',
        mode='lines',
        line=dict(width=0.5),
        stackgroup='one',
        name=asset
    ))
fig.update_layout(height=750, yaxis_range=(0, 100), legend=dict(orientation="h"))
fig.show()
```

```python
# normalise liabilities by total liabilities 
for liability in liabilities:
    df_sivb_blsh[liability] = (df_sivb_blsh[liability] / df_sivb_blsh["Total liabilities"])*100

fig = go.Figure()
for liability in liabilities:
    fig.add_trace(go.Scatter(
        x=df_sivb_blsh["Year Ending"], y=df_sivb_blsh[liability],
        hoverinfo='x+y',
        mode='lines',
        line=dict(width=0.5),
        stackgroup='one',
        name=liability
    ))
fig.update_layout(height=600, yaxis_range=(0, 100), legend=dict(orientation="h"))
fig.show()
```
