```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Set the stock symbol for Silicon Valley Bank
symbol = "SIVB"

# Set the start and end dates for the stock price data
start_date = "1982-01-01"
end_date = "2023-03-12"

# Retrieve the stock price data for SVB
data = yf.download(symbol, start=start_date, end=end_date)

# Select the "Close" column from the data and plot it
close_prices = data["Close"]

plt.figure(figsize=(12,6))
plt.plot(close_prices)
plt.title("SVB Stock Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.yscale('log')
plt.show()
```
