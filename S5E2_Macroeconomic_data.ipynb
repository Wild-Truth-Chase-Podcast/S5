```python
import pandas as pd
import requests
import plotly.express as px

# Set FRED API key and series ID for GDP growth rate
fred_api_key = os.environ.get("FRED_API_KEY")
```

```python
# Define function to fetch GDP growth rate data from FRED API
def fetch_gdp_growth_rate_data():
    series_id = 'GDPC1'
    url = f'https://api.stlouisfed.org/fred/series/observations?series_id={series_id}&api_key={fred_api_key}&file_type=json'

    data = requests.get(url).json()

    df = pd.DataFrame(data['observations'])
    df['date'] = pd.to_datetime(df['date'])
    df['value'] = pd.to_numeric(df['value'])

    return df

# Call the function to fetch GDP growth rate data
gdp_growth_rate_df = fetch_gdp_growth_rate_data()

# Plot the GDP growth rate data using Plotly
fig = px.line(gdp_growth_rate_df, x='date', y='value', title='US GDP Growth Rate')
fig.show()
```

```python
# Plot the GDP growth rate data using Plotly
from datetime import datetime

fig = px.line(gdp_growth_rate_df, x='date', y='value', title='US GDP Growth Rate')
fig.update_xaxes(range = [datetime(2010, 1, 1), datetime(2024, 1, 1)])
fig.update_yaxes(range = [15_000, 22_000])
fig.show()
```

```python
# Define function to fetch inflation rate data from FRED API
def fetch_inflation_rate_data():
    series_id = 'CPALTT01USM657N'
    url = f'https://api.stlouisfed.org/fred/series/observations?series_id={series_id}&api_key={fred_api_key}&file_type=json'

    data = requests.get(url).json()

    df = pd.DataFrame(data['observations'])
    df['date'] = pd.to_datetime(df['date'])
    df['value'] = pd.to_numeric(df['value'], errors='coerce')

    return df

# Call the function to fetch inflation rate data
inflation_rate_df = fetch_inflation_rate_data()

# Plot the inflation rate data using Plotly
fig = px.line(inflation_rate_df, x='date', y='value', title='US Inflation Rate')
fig.show()
```
