```python
from google.colab import drive
import os

drive.mount('/content/drive')

# Load environment variables from file
with open('/content/drive/My Drive/env_vars.txt', 'r') as f:
    for line in f:
        name, value = line.strip().split('=')
        os.environ[name] = value
```

```python
import pandas as pd
import requests
import plotly.express as px

# Set FRED API key and series ID for GDP growth rate
fred_api_key = os.environ.get("FRED_API_KEY")
series_id = 'GDPC1'

# Define function to fetch GDP growth rate data from FRED API
def fetch_gdp_growth_rate_data():
    url = f'https://api.stlouisfed.org/fred/series/observations?series_id={series_id}&api_key={fred_api_key}&file_type=json'

    data = requests.get(url).json()

    df = pd.DataFrame(data['observations'])
    df['date'] = pd.to_datetime(df['date'])
    df['value'] = pd.to_numeric(df['value'])

    return df

# Call the function to fetch GDP growth rate data
gdp_growth_rate_df = fetch_gdp_growth_rate_data()

# Plot the GDP growth rate data using Plotly
fig = px.line(gdp_growth_rate_df, x='date', y='value', title='US GDP Growth Rate')
fig.show()
```
